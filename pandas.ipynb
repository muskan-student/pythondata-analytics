```python
fig,ax = plt.subplots(figsize =(15,6))
sns.barplot(x= 'country',y = 'Total', data = pf.head(10),ax = ax)
plt.title("top 10 countries with highest immigration")
plt.xticks(rotation = 90)
ax.axhline(pf.head(10)['Total'].mean(),color ='r',linestyle = "--")
ax.text(5 ,pf.head(10)['Total'].mean()+1000,pf.head(10)['Total'].mean())
ax.axhspan(300000,400000 ,color = 'green',alpha= .2)
plt.show()
```
