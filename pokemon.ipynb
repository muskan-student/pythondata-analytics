```python

import os 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('pokemon.csv')
df
```

```python
df['Name'].tolist()
```

```python
#fig,ax =plt.subplots(figsize =(15,6))
sns.catplot(
    data =df,
    x = 'Type 1',
    y = 'Attack',
    hue = 'Legendary',
    kind = 'box'
)
```

```python
ax = sns.countplot(
    data= df,
    x= 'Generation')
ax.set_title('Pokemon per generations ')
ax.set(xlabel = 'Generation', ylabel= 'count')    

```

```python
 fig , ax =plt.subplots(figsize  =(15,6))
 sns.scatterplot(
    data = df,
    x = 'Attack',y = 'Defense',
    alpha = 1,color= 'red',hue = 'Type 1'
)
plt.xlabel('Attack')
plt.ylabel ('Defense')
```

```python
fig,ax = plt.subplots(figsize = (15,6))
sns.scatterplot(
    data = df,
    x = 'Speed',y ='Defense',
    alpha= 1 ,hue = 'Type 2'
)
plt.xlabel('Speed')
plt.ylabel('Defense')
```

```python
sns.lineplot(
    data = df,
    x = 'Speed',
    kind = 'line',
    alpha= 1,linestyle ='--',
    color = 'green'
)
sns.lineplot(
    data = df,
    y = 'Defence',
    kind = 'line',
    linestyle = ':',alpha = 1,
    color = 'red'

)
plt.xlabel = 'x-axis',
plt.ylabel = 'y-axis',
plt.title = 'Line plot'
plt.show()
```
